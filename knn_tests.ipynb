```python
from sklearn import neighbors, datasets
from sklearn.neural_network import MLPClassifier
import numpy as np

#Carrega a base de dados
dados1 = datasets.load_breast_cancer()
dados2 = datasets.load_digits()
dados3 = datasets.load_iris()

#Cria um vetor aleatorio de mesma qtde de instancias da base de dados para criar conjuntos de treinamento aleatorio
perm1 = np.random.permutation(dados1.target.size)
perm2 = np.random.permutation(dados2.target.size)
perm3 = np.random.permutation(dados3.target.size)

#Cria dados para treinamento e testes escolhidos aleatoriamente
dados_cancer = dados1.data[perm1]
dados_cancer_rotulos = dados1.target[perm1] 
##
dados_digits = dados2.data[perm2]
dados_digits_rotulos = dados2.target[perm2]
##
dados_iris = dados3.data[perm3]
dados_iris_rotulos = dados3.target[perm3]
##

dados_cancer_treina = dados_cancer[:len(dados_cancer)//2]
dados_cancer_treina_rotulos = dados_cancer_rotulos[:len(dados_cancer_rotulos)//2]
##
dados_digits_treina = dados_digits[:len(dados_digits)//2]
dados_digits_treina_rotulos = dados_digits_rotulos[:len(dados_digits_rotulos)//2]
##
dados_iris_treina = dados_iris[:len(dados_iris)//2]
dados_iris_treina_rotulos = dados_iris_rotulos[:len(dados_iris_rotulos)//2]

dados_cancer_teste = dados_cancer[len(dados_cancer)//2:]
dados_cancer_teste_rotulos = dados_cancer_rotulos[len(dados_cancer_rotulos)//2:]
##
dados_digits_teste = dados_digits[len(dados_digits)//2:]
dados_digits_teste_rotulos = dados_digits_rotulos[len(dados_digits_rotulos)//2:]
##
dados_iris_teste = dados_iris[len(dados_iris)//2:]
dados_iris_teste_rotulos = dados_iris_rotulos[len(dados_iris_rotulos)//2:]


#Algoritmo KNN e MLP
knn_cancer = neighbors.KNeighborsClassifier()
knn_digits = neighbors.KNeighborsClassifier()
knn_iris = neighbors.KNeighborsClassifier()
mlp_cancer = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(100,),random_state=3)
mlp_digits = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(100,),random_state=3)
mlp_iris = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(100,),random_state=3)

#Treinamento
knn_cancer.fit(dados_cancer_treina, dados_cancer_treina_rotulos)
knn_digits.fit(dados_digits_treina, dados_digits_treina_rotulos)
knn_iris.fit(dados_iris_treina, dados_iris_treina_rotulos)

mlp_cancer.fit(dados_cancer_treina, dados_cancer_treina_rotulos)
mlp_digits.fit(dados_digits_treina, dados_digits_treina_rotulos)
mlp_iris.fit(dados_iris_treina, dados_iris_treina_rotulos)

#Avaliacao

print ("Accuracy knn_cancer: ",knn_cancer.score(dados_cancer_teste, dados_cancer_teste_rotulos))
print ("Accuracy knn_digits: ",knn_digits.score(dados_digits_teste, dados_digits_teste_rotulos))
print ("Accuracy knn_iris: ",knn_iris.score(dados_iris_teste, dados_iris_teste_rotulos))

print ("Accuracy mlp_cancer: ",mlp_cancer.score(dados_cancer_teste, dados_cancer_teste_rotulos))
print ("Accuracy mlp_digits: ",mlp_digits.score(dados_digits_teste, dados_digits_teste_rotulos))
print ("Accuracy mlp_iris: ",mlp_iris.score(dados_iris_teste, dados_iris_teste_rotulos))
```

```python
from sklearn import neighbors, datasets
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split, cross_val_score, ShuffleSplit
import numpy as np

#Carrega a base de dados
dados1 = datasets.load_breast_cancer()
dados2 = datasets.load_digits()
dados3 = datasets.load_iris()

#Partição
cv = ShuffleSplit (n_splits=5, test_size=0.4)

#Algoritmo KNN
knn_cancer = neighbors.KNeighborsClassifier()
knn_digits = neighbors.KNeighborsClassifier()
knn_iris = neighbors.KNeighborsClassifier()

#Cross Validation KNN 
scores_knn_cancer = cross_val_score(knn_cancer, dados1.data, dados1.target, cv=cv)
scores_knn_digits = cross_val_score(knn_digits, dados2.data, dados2.target, cv=cv)
scores_knn_iris = cross_val_score(knn_iris, dados3.data, dados3.target, cv=cv)

#Algoritmo MLP
mlp_cancer = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(100,),random_state=3)
mlp_digits = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(100,),random_state=3)
mlp_iris = MLPClassifier(solver='lbfgs',hidden_layer_sizes=(100,),random_state=3)

#Cross Validation MLP
scores_mlp_cancer = cross_val_score(mlp_cancer, dados1.data, dados1.target, cv=cv)
scores_mlp_digits = cross_val_score(mlp_digits, dados2.data, dados2.target, cv=cv)
scores_mlp_iris = cross_val_score(mlp_iris, dados3.data, dados3.target, cv=cv)

print (scores_knn_cancer)
print("Accuracy knn_cancer: %0.2f (+/- %0.2f)" % (scores_knn_cancer.mean(), scores_knn_cancer.std() * 2))

print (scores_knn_digits)
print("Accuracy knn_digits: %0.2f (+/- %0.2f)" % (scores_knn_digits.mean(), scores_knn_digits.std() * 2))

print (scores_knn_iris)
print("Accuracy knn_iris: %0.2f (+/- %0.2f)" % (scores_knn_iris.mean(), scores_knn_iris.std() * 2))

print (scores_mlp_cancer)
print ("Accuracy mlp_cancer: %0.2f (+/- %0.2f)" % (scores_mlp_cancer.mean(), scores_mlp_cancer.std() *2))

print (scores_mlp_digits)
print ("Accuracy mlp_digits: %0.2f (+/- %0.2f)" % (scores_mlp_digits.mean(), scores_mlp_digits.std() *2))

print (scores_mlp_iris)
print ("Accuracy mlp_iris: %0.2f (+/- %0.2f)" % (scores_mlp_iris.mean(), scores_mlp_iris.std() *2))
```
